#` Criando classes`

```
class Carro:
    def __init__(self, cor , qtd_portas,tipo_combustivel,potencia,qtd_combustivel,is_ligado,velocidade):
        self.cor = cor
        self.qtd_portas = qtd_portas
        self.tipo_combustivel = tipo_combustivel
        self.potencia = potencia
        self.qtd_combustivel = qtd_combustivel
        self.is_ligado = is_ligado
        self.velocidade = velocidade

    def  abastecer(self, valor):
        self.qtd_combustivel += valor


    def ligar (self):
        if self.is_ligado:
            print ('O carro já está ligado')
        else:    
            self.is_ligado = True

    def desligar(self):
        if self.is_ligado == True:
            print ('O carro já esta desligado')
        else:
            self.is_ligado = False  

    def  acelerar(self,velocidade):
        if self.is_ligado:
            self.velocidade += velocidade
        else:
            print ('O carro precisa está ligado')    



    def __str__(self):
        return f'Cor:{self.cor} - Qtd_Portas:{self.qtd_portas} - Combustível:{self.tipo_combustivel} \
- Potência: {self.potencia} - Qtd_combustível:{self.qtd_combustivel} - Ligado:{self.is_ligado} - Velocidade:{self.velocidade}'

meuCarro = Carro ('vermelho',4,'gasolina',1.0,0,False,0)
print (meuCarro)

meuCarro.ligar()
meuCarro.abastecer(50)
meuCarro.acelerar(50)

print (meuCarro)



```

```
numeros = list()
tamanho = int (input ('Digite o tamanho do vetor: '))
for i in range(tamanho):
    valor = int (input(f'Digite o número do vetor na posição {i}:--> '))
    numeros.append(valor)
print ('Vetor',numeros)
print ('Posição',numeros[1])   

pesquisar = int (input('Digite o número a  ser pesquisado no vetor: '))
posicao_resultado = -1

for i in range(tamanho):
    if numeros [i] == pesquisar:
        posicao_resultado = 1
        break
if posicao_resultado < 0:
    print('O elemento não foi encontrado no vetor')
else:
    print (f'O elemento encontrado na podição {posicao_resultado}')    

# Busca Linear
for i in range(tamanho):
    indice_menor =  i
    for j in range(int (i + 1) , tamanho):
        if numeros[j] < numeros[indice_menor]:
            indice_menor = j
    temp = numeros[indice_menor]
    numeros[indice_menor] = numeros[i]    
    numeros[i] = temp
    print ('Vetor:', numeros)

# Buscar Bínaria

resultado = -1
inicio = 0
fim = tamanho - 1
meio = 0
alvo = int (input('Digite o elemento a ser encontrado:'))
while inicio <= fim:
    meio = int ((inicio + fim) / 2)
    if (numeros[meio] < alvo):
        inicio = meio + 1
    elif (numeros[meio] > alvo):
        fim = meio - 1
    else:
        resultado = meio
        break

if resultado < 0:
    print ('O elemento não encontrado!')
else:
    print (f'O elemento {alvo} está na posição {resultado}')                    


```

```
 #-Inserindo elementos em uma posição específica.
 
 class Vetor:
    def __init__(self, tamanho):
        self.__tamanho = tamanho
        self.__elemento = [None] * tamanho
        self.__posicao = 0
 
    def inserir_elemento_posicao(self,elemento,posicao):
        vetor_inicio = self.__elemento [:posicao] + [None]
        vetor_final = self.__elemento [posicao:]
        vetor_inicio [len(vetor_inicio) - 1] = elemento
        self.__elemento = vetor_inicio + vetor_final
        self.__posicao += 1
        

    def inserir_elemento_final(self,elemento):
        if self.__posicao >= len(self.__elemento):
            self.__elemento = self.__elemento +[None]

        self.__elemento[self.__posicao] = elemento
        self.__posicao += 1

        self.__elemento[posicao] = elemento
    
    def inserir_elemento_final(self,elemento):
        if self.__posicao >= len(self.__elemento):
            self.__elemento = self.__elemento +[None]

        self.__elemento[self.__posicao] = elemento
        self.__posicao += 1
 

    def  listar_elemento(self,posicao):
        return self.__elemento[posicao]
 
 
 
 
print (30 * '-' , 'MENU' , 30 * '-')
print('1. -> Vetores')
print('2. ->  Listas Ligadas')
 
 
menu = int (input ('Digite a posição desejada!: -> '))
if menu == 1:
    vetor_teste = Vetor(3)
    vetor_teste.inserir_elemento_final(1)
    vetor_teste.inserir_elemento_final(2)
    vetor_teste.inserir_elemento_final(3)
    print (vetor_teste.listar_elemento(0))
    print (vetor_teste.listar_elemento(1))
    print (vetor_teste.listar_elemento(2))
   
```
